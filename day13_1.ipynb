```python
import numpy as np
v1=np.array([2,4,6,8])
v2=np.array([1,3,5,7])
print("Addition:",v1+v2)
print("Substraction:",v1-v2)
print("Multiplaction:",v1*v2)
print("Dot product:",np.dot(v1,v2))
```

```python
import numpy as np
a=np.random.randint(1,101,(4,4))
print("Array:",a)
print("Shape:",a.shape)
print("Dimenstion:",a.ndim)
print("Size:",a.size)
print("Type:",a.dtype)
print("Minimum:",a.min())
print("Maximum:",a.max())
```

```python
import numpy as np
a=np.random.randint(1,51,20)
b=a.reshape(4,5)
print("Matrix:",b)
print("Sum:",b.sum())
print("Mean:",b.mean())
print("Standard Deviation:",b.std())
print("Maximum value:",b.max(axis=1))
```

```python
import numpy as np
#input
a=int(input("Enter a numbers:"))
#Random numbers
b=np.random.randint(10,100,a)
print("Array:",b)
#Statistics
print("Mean:",np.mean(b))
print("Median:",np.median(b))
print("Standard deviation:",np.std(b))
print("Minimum:",np.min(b))
print("Maximum:",np.max(b))
#Reshape if possible
if a % 2 == 0:
  c = b.reshape(2, a//2)
  print("Reshaped array:",c)
  print("Sum:",c.sum(axis=1))
else:
  print("Cannot Reshape!")
```
