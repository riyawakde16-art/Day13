```python
import numpy as np
# 10 students, 5 subjects
m = np.random.randint(30, 101, (10, 5))
print("Marks:\n", m)
# Total and average
t = m.sum(axis=1)
a = m.mean(axis=1)
print("Total marks:", t)
print("Average marks:", a)
# Highest & lowest
print("Top student index:", t.argmax())
print("Lowest student index:", t.argmin())
# Class statistics
print("Class mean:", m.mean())
print("Class std:", m.std())
# Top 3 students marks
top3 = np.argsort(t)[-3:]
print("Top 3 students:\n", m[top3])
```
